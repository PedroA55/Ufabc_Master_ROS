```python
import bagpy
from bagpy import bagreader
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_name = 'dados'
#b = bagreader('../bags/dados.bag')
b = bagreader('../bags/' + file_name + '.bag')
```

```python
b.topic_table
```

```python
des_traj = b.message_by_topic(topic='/desired_trajectory')
des_traj_df = pd.read_csv(des_traj)
des_traj_df
```

```python
list(des_traj_df)
```

```python
actual_pos = b.message_by_topic(topic='/quad/kf/position')
actual_pos_df = pd.read_csv(actual_pos)
actual_pos_df
```

```python
# Posições
plt.subplot(3,1,1)
plt.title('Posições XYZ')
plt.plot(des_traj_df['Time'], des_traj_df['cartesian_point.pose.position.x'])
plt.plot(actual_pos_df['Time'], actual_pos_df['x'])
frame1 = plt.gca()
frame1.axes.xaxis.set_ticklabels([])
#frame1.axes.get_xaxis().set_ticks([])
#frame1.axes.get_xaxis().set_visible(False)
plt.grid()
plt.legend(['x(t) desejado','x(t) FK'])
plt.subplot(3,1,2)
plt.plot(des_traj_df['Time'], des_traj_df['cartesian_point.pose.position.y'])
plt.plot(actual_pos_df['Time'], actual_pos_df['y'])
plt.legend(['y(t) desejado','y(t) FK'])
frame2 = plt.gca()
frame2.axes.xaxis.set_ticklabels([])
#frame2.axes.get_xaxis().set_visible(False)
#frame2.axes.get_xaxis().set_visible(False)
plt.grid()
plt.subplot(3,1,3)
plt.plot(des_traj_df['Time'], des_traj_df['cartesian_point.pose.position.z'])
plt.plot(actual_pos_df['Time'], actual_pos_df['z'])
plt.legend(['z(t) desejado','z(t) FK'])
plt.grid()
plt.xlabel('tempo')
plt.show()
```

```python
# velocidades lineares
actual_vel = b.message_by_topic(topic='/quad/kf/vel')
actual_vel_df = pd.read_csv(actual_vel)
actual_vel_df
```

```python
# Velocidades
plt.subplot(3,1,1)
plt.title('Velocidades XYZ')
plt.plot(des_traj_df['Time'], des_traj_df['cartesian_point.velocity.linear.x'])
plt.plot(actual_vel_df['Time'], actual_vel_df['x'])
plt.legend(['$v_x$(t) desejado','$v_x$(t) FK'])
frame1 = plt.gca()
frame1.axes.xaxis.set_ticklabels([])
#frame1.axes.get_xaxis().set_ticks([])
#frame1.axes.get_xaxis().set_visible(False)
plt.grid()

plt.subplot(3,1,2)
plt.plot(des_traj_df['Time'], des_traj_df['cartesian_point.velocity.linear.y'])
plt.plot(actual_vel_df['Time'], actual_vel_df['y'])
plt.legend(['$v_y$(t) desejado','$v_y$(t) FK'])
frame2 = plt.gca()
frame2.axes.xaxis.set_ticklabels([])
#frame2.axes.get_xaxis().set_visible(False)
#frame2.axes.get_xaxis().set_visible(False)
plt.grid()

plt.subplot(3,1,3)
plt.plot(des_traj_df['Time'], des_traj_df['cartesian_point.velocity.linear.z'])
plt.plot(actual_vel_df['Time'], actual_vel_df['z'])
plt.legend(['$v_z$(t) desejado','$v_z$(t) FK'])
plt.grid()
plt.xlabel('tempo')
plt.show()
```

```python
actual_ati = b.message_by_topic(topic='/quad/kf/attitude')
actual_ati_df = pd.read_csv(actual_ati)
actual_ati_df
```

```python
# Pose
plt.subplot(2,2,1)
plt.title('Atitude Quaternions')
plt.plot(des_traj_df['Time'], des_traj_df['cartesian_point.pose.orientation.w'])
plt.plot(actual_ati_df['Time'], actual_ati_df['w'])
plt.legend(['w(t) desejado','w(t) FK'])
frame1 = plt.gca()
frame1.axes.xaxis.set_ticklabels([])
#frame1.axes.get_xaxis().set_ticks([])
#frame1.axes.get_xaxis().set_visible(False)
plt.grid()

plt.subplot(2,2,2)
plt.plot(des_traj_df['Time'], des_traj_df['cartesian_point.pose.orientation.x'])
plt.plot(actual_ati_df['Time'], actual_ati_df['x'])
plt.legend(['x(t) desejado','x(t) FK'])
frame2 = plt.gca()
frame2.axes.xaxis.set_ticklabels([])
#frame2.axes.get_xaxis().set_visible(False)
#frame2.axes.get_xaxis().set_visible(False)
plt.grid()

plt.subplot(2,2,3)
plt.plot(des_traj_df['Time'], des_traj_df['cartesian_point.pose.orientation.y'])
plt.plot(actual_ati_df['Time'], actual_ati_df['y'])
plt.legend(['y(t) desejado','y(t) FK'])
plt.grid()
plt.xlabel('tempo')

plt.subplot(2,2,4)
plt.plot(des_traj_df['Time'], des_traj_df['cartesian_point.pose.orientation.z'])
plt.plot(actual_ati_df['Time'], actual_ati_df['z'])
plt.legend(['z(t) desejado','z(t) FK'])
plt.grid()
plt.xlabel('tempo')

plt.show()
```

```python
actual_ati_euler = b.message_by_topic(topic='/quad/kf/euler')
actual_ati_euler_df = pd.read_csv(actual_ati_euler)
actual_ati_euler_df
```
